```python
# using with open() will automatically closes the file after we perform all the operations in it rather than opening a file manually
# opening a file in w mode will create a new file or override the already created file and create a new file
# opening a file in a mode allows use to append something to the end of the file and create a new file if that file does not exist
# function is a executable reusable block of code 
# pass keyword does nothing and its just there as a placeholder so that you make any indentation errors
# \n is used to point the cursor to the next line after finishing a line
# .readlines() will read all the lines from a text file
# .rstrip() will strip off the carridge return or we can say that it will remove the \n from our output so it doesnt point to the next line
# .split() it will split the string when it encounters the passed character and returns a list
# to do encryption of the password we use the module called as CRYPTOGRAPHY and we have to install it on our commad line using pip install cryptography
# from cryptography.fernet import Fernet --> this is the module that is going to allow us to encrypt text essentially
# first thing we need to do to use the above module is to define a key. we combine this key with the master password to encrypt our text.
# so the master password in combination with the key will be able to encrypt and decrypt our text and if we type the wrong master password we will get the worng decrypted text
# Fernet.generate_key() --> this function is used to generate a key
# key.key --> this is how we create a key file rather than a text file
#Fernet(key_name) --> this is used to initialize the encryption module
# fer.encrypt(pwd.encode()) --> this is used to convert the password into its encrypted version
# .encode() --> it will takes the string and turns it into bytes.
# b'hello' this is considered as a byte string in python rather than a regular string 'hello'
# .decode() this function is going to take the byte string and convert it into a regular string


from cryptography.fernet import Fernet


# def write_key():   # this function is used to create a key for us and store it in a file
#     key = Fernet.generate_key()
#     with open('key.key', 'wb') as key_file:
#         key_file.write(key)      
# write_key()


def load_key():      # this function is used to load the already existing key 
    file = open('key.key', 'rb')
    key = file.read()
    file.close()
    return key

   
master_pwd = input("What is the master password? ")

key = load_key() + master_pwd.encode()  # we were adding the master password with the key and then using that as an actual key
fer = Fernet(key)


def view():  
    with open('passwords.txt', mode='r') as f:        
        for line in f.readlines():
            data = line.rstrip()
            user, passw = data.split("|")
            print("User:", user, ", Password:", fer.decrypt(passw.encode()))  # this is used to decrypt the encrypted password
             

def add():
    name = input("Account name: ")
    pwd = input("Password: ")  

    with open('passwords.txt', mode='a') as f:        
        f.write(name + " | "+ fer.encrypt(pwd.encode()).decode() + "\n")   # it will convert the string into bytes and then encrypt the key
        
        
while True:
    mode = input("would you like to add a new password or view existing ones (view/add), press q to quit? ").lower()
    
    if mode == "q":
        break

    if mode == "view":
        view()
    elif mode == "add":
        add()
    else:
        print("Invalid mode!!")
        continue
        
    
```
